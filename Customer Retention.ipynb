### This column directly gives us an idea on customer behaviour. We need to observe the answers of customers who are using online shopping platform over 4 years to get an idea as to what are the factors that worked favourable for them as well as the customers who are using the platform for 4 years or less to understand what are the scenarios which did not work for them

```python
Data['5 Since How Long You are Shopping Online ?']=Data['5 Since How Long You are Shopping Online ?'].replace({'Less than 1 year':0,'1-2 years':1,'2-3 years':2,'3-4 years':3,'Above 4 years':4})
```

```python
Data['6 How many times you have made an online purchase in the past 1 year?'].unique()
```

```python
sns.catplot(x='6 How many times you have made an online purchase in the past 1 year?',y='5 Since How Long You are Shopping Online ?',data=Data,kind='point',height=8,aspect=2)
```

### Few observations derived from the above pattern:

### 1. Maximum online platform has been visited by customers who are using this for 0-2 years, interestingly it is between this same period that the lowest online visit has been found (1.5 to 2 years).

### 2. People who have been using the platform for 3 years and above have also visited between 21-30 and 30-40 times.

### 3. Therefore it can also be concluded that if a customer stays over 2.5 years there is a high chance of retention for that customer
