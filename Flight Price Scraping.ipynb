```python
import selenium
import time
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import time #importing necessary libraries
```

```python
driver=webdriver.Chrome(r"C:\Webdriver\chromedriver.exe") #loading chrome from C-drive and assigning a variable to it
```

```python
driver.get("https://flight.yatra.com/air-search-ui/dom2/trigger?ADT=1&CHD=0&INF=0&class=Economy&destination=BOM&destinationCountry=IN&flexi=0&flight_depart_date=01%2F11%2F2022&hb=&noOfSegments=1&origin=DEL&originCountry=IN&type=O&unique=1667123425416&viewName=normal") # loading website in the chrome driver
```

```python
Name=[]
Name_tag=driver.find_elements_by_xpath("//span[@class='i-b text ellipsis']")
for i in Name_tag:
    Name.append(i.text)
    
Name
```

```python
Names=pd.DataFrame(Name)
```

```python
Depart_Time=[]
Depart_tag=driver.find_elements_by_xpath("//div[@class='i-b pr']")
for i in Depart_tag:
    Depart_Time.append(i.text)
    
Depart_Time
```

```python
Depart_Times=pd.DataFrame(Depart_Time)
```

```python
Arrival_Time=[]
Arrival_tag=driver.find_elements_by_xpath("//p[@class='bold fs-15 mb-2 pr time']")
for i in Arrival_tag:
    Arrival_Time.append(i.text.split('\n')[0])
    
Arrival_Time
```

```python
Arrival_Times=pd.DataFrame(Arrival_Time)
```

```python
Duration=[]
Duration_tag=driver.find_elements_by_xpath("//div[@class='stop-cont pl-13']")
for i in Duration_tag:
    Duration.append(i.text.split('\n')[0])
    
Duration
```

```python
Durations=pd.DataFrame(Duration)
```

```python
Price=[]
Price_tag=driver.find_elements_by_xpath("//div[@class='i-b tipsy fare-summary-tooltip fs-18']")
for i in Price_tag:
    Price.append(i.text)
    
Price
```

```python
Prices=pd.DataFrame(Price)
```

```python
DF1=pd.concat([Names,Depart_Times,Arrival_Times,Durations,Prices],axis=1)
DF1.columns=['Name','Depart_Time','Arrival_Time','Duration','Price']
DF1
```

```python
DF1['Date']=['31/10']*25
```
