# ODI Teams

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
page=requests.get('https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
Team=[]
for i in soup.find('td',class_="rankings-block__banner--team-name"):
    Team.append(i.text.replace('\n',''))
    
for i in soup.find_all('td',class_="table-body__cell rankings-table__team"):
    Team.append(i.text.replace('\n',''))
```

```python
Team 
```

```python
for ele in Team:
    if ele=='':
        Team.remove(ele)
    else:
        pass
```

```python
Team
```

```python
Team1=Team[0]+Team[1]
```

```python
Teams1=Team1.split()
```

```python
del Team[0:2]
```

```python
Teams=Teams1+Team
```

```python
Teams
```

```python
Women_Team=pd.DataFrame(Teams)
```

```python
Matches=[]
for i in soup.find('td',class_="rankings-block__banner--matches"):
    Matches.append(i.text.replace('\n',''))
    
for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    Matches.append(i.text.replace('\n',''))
```

```python
Matches
```

```python
Matches1=Matches[0]
Matches0=Matches1.split()
```

```python
Matches2=Matches[1:21:2]
```

```python
Match=Matches0+Matches2
```

```python
Match1=pd.DataFrame(Match)
```

```python
Point=[]
for i in soup.find('td',class_="rankings-block__banner--points"):
    Point.append(i.text.replace('\n',''))
for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    Point.append(i.text.replace('\n',''))
```

```python
Point
```

```python
Points=Point[0:21:2]
```

```python
Points1=pd.DataFrame(Points)
```

```python
Rating=[]
for i in soup.find('td',class_="rankings-block__banner--rating u-text-right"):
    Rating.append(i.text.replace('\n',''))
for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    Rating.append(i.text.replace('\n',''))
```

```python
Rating
```

```python
for ele in Rating:
    if ele=='':
        Rating.remove(ele)
    else:
        pass
```

```python
Rating
```

```python
Rating1=pd.DataFrame(Rating)
```

```python
ODI_Team=pd.concat([Women_Team,Match1,Points1,Rating1],axis=1)
ODI_Team.columns=["Team","Matches","Points","Ratings"]
```

```python
ODI_Team.head(10)
```

# ODI Batting

```python
page=requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
Batsmen=[]
for i in soup.find_all('div',class_="rankings-block__banner--name-large"):
    Batsmen.append(i.text.replace('\n',''))
for i in soup.find_all('td',class_="table-body__cell rankings-table__name name"):
    Batsmen.append(i.text.replace('\n',''))
```

```python
Batsmen
```

```python
Batsmen1=pd.DataFrame(Batsmen)
```

```python
Team=[]
for i in soup.find_all('div',class_="rankings-block__banner--nationality"):
    Team.append(i.text.replace('\n',''))
for i in soup.find_all('span',class_="table-body__logo-text"):
    Team.append(i.text.replace('\n',''))
```

```python
Team
```

```python
Team1=pd.DataFrame(Team)
```

```python
Rating=[]
for i in soup.find_all('div',class_="rankings-block__banner--rating"):
    Rating.append(i.text.replace('\n',''))
for i in soup.find_all('td',class_="table-body__cell rating"):
    Rating.append(i.text.replace('\n',''))
```

```python
Rating
```

```python
Rating1=pd.DataFrame(Rating)
```

```python
ODI_Batting=pd.concat([Batsmen1,Team1,Rating1],axis=1)
ODI_Batting.columns=["Batsmen","Team","Rating"]
```

```python
ODI_Batting.head(10)
```
