```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException , StaleElementReferenceException
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome(r"C:\Webdriver\chromedriver.exe") #loading chrome from C-drive and assigning a variable to it
```

# Scraping Headphones

```python
# Getting the website to driver
driver.get('https://www.flipkart.com/boult-audio-probass-curve-neckband-bluetooth-headset/product-reviews/itm3e1f7accb87d7?pid=ACCFEZ99MFMDR8CM&lid=LSTACCFEZ99MFMDR8CMAJFROP&marketplace=FLIPKART')
```

```python
Review=[]
for i in range(0,200):
    Review_tag=driver.find_elements_by_xpath("//p[@class='_2-N8zT']")
    try:
        for i in Review_tag:
            Review.append(i.text)
    except NoSuchElementException:
        Review.append("Not Available")
        next_button=driver.find_element_by_xpath("/html/body/div[1]/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[11]").click()
        time.sleep(10)
```

```python
Review
```

```python
len(Review)
```

```python
Rating=[]
for i in range(0,200):
    Rating_tag=driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']")
    try:
        for i in Rating_tag:
            Rating.append(i.text.split('\n')[0])
    except NoSuchElementException:
        Rating.append("Not Available")
        next_button=driver.find_element_by_xpath("/html/body/div/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[11]").click()
        time.sleep(10)
```

```python
len(Rating)
```

```python
#Saving in dataframe
Headphones=pd.DataFrame({'Product_Review':Review,'Ratings':Rating})
```

```python
Headphones
```
